```python
import json
fp = open("test1.json","r")
content = fp.read()
d = json.loads(content)

d['glossary']['title'] = "New title for JSON file"
# print(d['glossary']['GlossDiv']['GlossList']['GlossEntry']['ID'])
d['glossary']['GlossDiv']['GlossList']['GlossEntry']['ID'] = ("SGML","HTML")

j = json.dumps(d,indent = 4 )

fp.close()

fp = open("test2.json","w")
fp.write(j)
fp.close()
```

```python
def gen_dict(d,d2):
    for key,value in d.items():
        if type(value) == dict:
            return gen_dict(value,d2)
        else:
            d2[key] = value
    return d2


d = {
    "glossary": {
        "title": "example glossary",
		"GlossDiv": {
            "title": "S",
			"GlossList": {
                "GlossEntry": {
                    "ID": "SGML",
					"SortAs": "SGML",

					"GlossSee": "markup"
                }
            }
        }
    }
}

d2 = {}
print(d)
result = gen_dict(d,d2)
print(result)
```

```python
import xmltodict

fp = open("test1.xml","r")
content = fp.read()
# print(content)
fp.close()

d = xmltodict.parse(content)
# print(d)

d['glossary']['title'] = "New title for XML"
# print(d)
d['glossary']['GlossDiv']['GlossList']['GlossEntry']['@ID'] = "HTML"

x = xmltodict.unparse(d)
print(x)

fp = open("test2.xml","w")
fp.write(x)
fp.close()
```
